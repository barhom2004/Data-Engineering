Multiple Inheritance

الشرح

الـ Multiple Inheritance معناها إن الكلاس يقدر يرث من أكثر من كلاس في نفس الوقت.

يعني بدل ما يرث من أب واحد بس، يقدر يأخذ دوال وخصائص من أكثر من كلاس.
هذا يفيد لما يكون عندك خصائص مختلفة وتريد تجمعهم في كلاس واحد بدون ما تكرر الكود.

بايثون لما يبحث عن دالة، يستخدم ترتيب معين اسمه
MRO (Method Resolution Order)
عشان يعرف يبدأ من فين.

الفكرة حلوة وتختصر كود،
بس لازم تستخدمها بعقل عشان ما يصير تعقيد في المشروع.

```python
class Car:
    def __init__(self, brand, speed):
        self.brand = brand
        self.speed = speed

    def drive(self):
        print(f"{self.brand} is driving at {self.speed} km/h")


class Electric:
    def __init__(self, battery_level):
        self.battery_level = battery_level

    def charge(self):
        print(f"Battery level is {self.battery_level}%")


class ElectricCar(Car, Electric):
    def __init__(self, brand, speed, battery_level, range_km):
        Car.__init__(self, brand, speed)
        Electric.__init__(self, battery_level)
        self.range_km = range_km

    def show_info(self):
        self.drive()
        self.charge()
        print(f"Range: {self.range_km} km")


tesla = ElectricCar("Tesla", 120, 85, 400)
tesla.show_info()
```
