```python
import numpy as np


def text_to_vector(text, bag_of_words):
    words = text.split()
    vector = []
    
    for word in bag_of_words:
        if word in words:
            vector.append(1)
        else:
            vector.append(0)
    
    return np.array(vector)



text1 = "AI Python Data"
text2 = "AI Model"

bag_of_words = ["AI", "Python", "Data", "Model"]

v1 = text_to_vector(text1, bag_of_words)
v2 = text_to_vector(text2, bag_of_words)

print("Vector 1:", v1)
print("Vector 2:", v2)


dot_product = np.dot(v1, v2)


norm1 = np.linalg.norm(v1)
norm2 = np.linalg.norm(v2)


cosine_similarity = dot_product / (norm1 * norm2)

print("Cosine Similarity:", cosine_similarity)
```
